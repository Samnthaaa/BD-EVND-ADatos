# La Sentencia Break y Continue

break - Sale del bucle inmediatamente, e incondicional termina la operacion del bucle, el programa empieza a ejecutar lq instruccion mas cercana despues del cuerpo del bucle.

Continue -Se comporta como si el programa hubiera llegado repentinamente al final del cuerpo; el siguiente turno se inicio y la expresion de condicion se prueba de inmediato.

```python

print("La instruccion break")
for i in range (1,6):
    print('Dentro del bucle')
    if i ==3:
        break

print ('Salte del Bucle')
```

```python
print('\nLa instruccion')
for i in range (1, 6):
    if i == 3:
        continue
    print(f'dentro del bucle. {i}')
print('Fuera del Bucle')
```
